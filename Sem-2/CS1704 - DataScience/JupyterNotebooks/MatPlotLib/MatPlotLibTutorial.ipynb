```python
qty = np.array([10,30,40,20])
fruits = ['apple','banana','orange','peach']
exploding = [0.5,0,0,0]
colors = ['red','orange','yellow','purple']
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)
plt.pie(qty,labels=fruits,colors=colors,explode=exploding,shadow=True,startangle=270,rotatelabels=True)
plt.legend(title='Fruits',loc ='lower left')
plt.show()
```
